```python
# Colab-ready code to demonstrate activation functions
import numpy as np
import matplotlib.pyplot as plt

# Activation functions with definitions
def sigmoid(x):
    """Sigmoid: Squashes input to range (0, 1); used for binary classification."""
    return 1 / (1 + np.exp(-x))

def tanh(x):
    """Tanh: Maps input to range (-1, 1); often used in hidden layers for better normalization."""
    return np.tanh(x)

def relu(x):
    """ReLU: Sets negative values to 0; widely used for its simplicity and efficiency."""
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.1):
    """Leaky ReLU: Allows small gradient for negative values; prevents dying neurons."""
    return np.where(x > 0, x, alpha * x)

def softmax(x):
    """Softmax: Converts input into probabilities summing to 1; used in multi-class classification."""
    exp_x = np.exp(x - np.max(x))  # Stability improvement
    return exp_x / exp_x.sum(axis=0)

# Generate input data
x = np.linspace(-10, 10, 100)

# Calculate outputs
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)
y_softmax = softmax(np.array([x, x * 0.5]))  # Example with 2 classes

# Plot graphs
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.plot(x, y_sigmoid, label="Sigmoid")
plt.title("Sigmoid")
plt.grid()
plt.legend()

plt.subplot(2, 3, 2)
plt.plot(x, y_tanh, label="Tanh", color="orange")
plt.title("Tanh")
plt.grid()
plt.legend()

plt.subplot(2, 3, 3)
plt.plot(x, y_relu, label="ReLU", color="green")
plt.title("ReLU")
plt.grid()
plt.legend()

plt.subplot(2, 3, 4)
plt.plot(x, y_leaky_relu, label="Leaky ReLU", color="red")
plt.title("Leaky ReLU")
plt.grid()
plt.legend()

plt.subplot(2, 3, 5)
plt.plot(x, y_softmax[0], label="Softmax Class 1", color="purple")
plt.plot(x, y_softmax[1], label="Softmax Class 2", color="brown")
plt.title("Softmax")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Colab-ready code to demonstrate activation functions
import numpy as np
import matplotlib.pyplot as plt

# Activation functions
def sigmoid(x):
    """Sigmoid: Maps inputs to a range between 0 and 1, making it ideal for binary classification."""
    return 1 / (1 + np.exp(-x))

def tanh(x):
    """Tanh: Maps inputs to a range between -1 and 1, often used in hidden layers for better normalization."""
    return np.tanh(x)

def relu(x):
    """ReLU: Converts negative inputs to zero while keeping positive inputs unchanged, introducing non-linearity."""
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.1):
    """Leaky ReLU: Addresses the dying ReLU problem by allowing a small gradient for negative inputs."""
    return np.where(x > 0, x, alpha * x)

def elu(x, alpha=1.0):
    """ELU: Extends ReLU by allowing smooth gradients for negative values to prevent dead neurons."""
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def swish(x):
    """Swish: Combines sigmoid and linear functions, retaining information for better gradient flow."""
    return x / (1 + np.exp(-x))

def softmax(x):
    """Softmax: Converts input into probabilities that sum to 1, used in multi-class classification problems."""
    exp_x = np.exp(x - np.max(x))  # Stability improvement
    return exp_x / exp_x.sum(axis=0)

# Generate input data
x = np.linspace(-10, 10, 100)

# Calculate outputs
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)
y_elu = elu(x)
y_swish = swish(x)
y_softmax = softmax(np.array([x, x * 0.5]))  # Example with 2 classes

# Plot graphs
plt.figure(figsize=(15, 10))

# Sigmoid
plt.subplot(3, 3, 1)
plt.plot(x, y_sigmoid, label="Sigmoid")
plt.title("Sigmoid")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.legend()

# Tanh
plt.subplot(3, 3, 2)
plt.plot(x, y_tanh, label="Tanh", color="orange")
plt.title("Tanh")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.legend()

# ReLU
plt.subplot(3, 3, 3)
plt.plot(x, y_relu, label="ReLU", color="green")
plt.title("ReLU")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.legend()

# Leaky ReLU
plt.subplot(3, 3, 4)
plt.plot(x, y_leaky_relu, label="Leaky ReLU", color="red")
plt.title("Leaky ReLU")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.legend()

# ELU
plt.subplot(3, 3, 5)
plt.plot(x, y_elu, label="ELU", color="blue")
plt.title("ELU")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.legend()

# Swish
plt.subplot(3, 3, 6)
plt.plot(x, y_swish, label="Swish", color="purple")
plt.title("Swish")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid()
plt.legend()

# Softmax
plt.subplot(3, 3, 7)
plt.plot(x, y_softmax[0], label="Softmax Class 1", color="purple")
plt.plot(x, y_softmax[1], label="Softmax Class 2", color="brown")
plt.title("Softmax")
plt.xlabel("Input")
plt.ylabel("Probability")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()

# Explanations
print("""
Activation Functions Explained:

1. Sigmoid:
   - Definition: Maps inputs to a range between 0 and 1, making it ideal for binary classification problems.
   - Graph: Displays a smooth curve showing probabilities for inputs.

2. Tanh:
   - Definition: Maps inputs to a range between -1 and 1, centering data for better normalization.
   - Graph: Symmetrical curve centered at zero.

3. ReLU:
   - Definition: Converts negative inputs to zero while keeping positive inputs unchanged, introducing non-linearity.
   - Graph: Outputs zero for negative inputs and linearly increases for positive inputs.

4. Leaky ReLU:
   - Definition: Addresses the dying ReLU problem by allowing a small gradient for negative inputs.
   - Graph: Similar to ReLU but with a small slope for negative values.

5. ELU:
   - Definition: Extends ReLU by allowing smooth gradients for negative values to prevent dead neurons.
   - Graph: Smooth transition for negative values and linear for positive.

6. Swish:
   - Definition: Combines sigmoid and linear functions, retaining information for better gradient flow.
   - Graph: Smooth curve, slightly better than ReLU for deep networks.

7. Softmax:
   - Definition: Converts input into probabilities that sum to 1, used in multi-class classification problems.
   - Graph: Displays probabilities for different classes as inputs vary.

Run the code above in Colab to generate these graphs and understand the behavior of each activation function interactively.
""")
```
