```python
# Importing the necessary modules from Keras
from keras.models import Sequential  # For creating a sequential model
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten  # Layers for the model

# Creating a Sequential model
model = Sequential([
    # Adding a Dense layer (fully connected layer)
    # This layer has 16 neurons, takes input of shape (30, 30, 3), and uses ReLU activation
    Dense(16, input_shape=(30, 30, 3), activation='relu'),

    # Adding a 2D Convolutional layer
    # This layer has 32 filters, each of size 4x4, uses ReLU activation, and 'valid' padding
    Conv2D(32, kernel_size=(4, 4), activation='relu', padding='valid'),

    # Adding a MaxPooling layer
    # Pool size is 2x2, stride is 2, and 'valid' padding is used (no extra padding)
    MaxPooling2D(pool_size=(2, 2), strides=2, padding='valid'),

    # Adding another Dense layer for the output
    # This layer has 2 neurons (e.g., for binary classification) and uses softmax activation
    Dense(2, activation='softmax')
])

# Printing a summary of the model
# This will show the layers, their output shapes, and the number of parameters
model.summary()
```
